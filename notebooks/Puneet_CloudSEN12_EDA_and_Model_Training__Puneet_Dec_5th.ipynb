```python
import os
import rasterio
import numpy as np
import matplotlib.pyplot as plt
from glob import glob

# 1. Define the base path to the extracted dataset and subdirectories
# The zip file extracted to /content/CloudSEN12_dataset/CloudSEN12_Dataset/
base_dataset_path = '/content/CloudSEN12_dataset/CloudSEN12_Dataset/'

train_path = os.path.join(base_dataset_path, 'train')
test_path = os.path.join(base_dataset_path, 'test')
val_path = os.path.join(base_dataset_path, 'val')

print(f"Base dataset path: {base_dataset_path}")
print(f"Train path: {train_path}")
print(f"Test path: {test_path}")
print(f"Val path: {val_path}")

# 3. Create an empty list to store image properties
image_properties = {
    'train': [],
    'test': [],
    'val': []
}

# Helper function to visualize an image
def visualize_image(image_path, title, ax):
    with rasterio.open(image_path) as src:
        img = src.read()

        # For multi-channel images, select first 3 bands for RGB display if available
        if img.shape[0] >= 3:
            # Permute to (H, W, C) for matplotlib
            display_img = np.transpose(img[:3, :, :], (1, 2, 0))

            # Normalize for visualization if not already in 0-255 range
            if display_img.max() > 255 or display_img.min() < 0:
                display_img = (display_img - display_img.min()) / (display_img.max() - display_img.min())

            ax.imshow(display_img)
        else:
            # For single or two channel images, display the first channel
            display_img = img[0, :, :]
            ax.imshow(display_img, cmap='gray')

        ax.set_title(title, fontsize=8)
        ax.axis('off')

# 4. For each dataset ('train', 'test', 'val'):
dataset_types = {'train': train_path, 'test': test_path, 'val': val_path}
sample_size = 3 # Number of images to sample from each set

plt.figure(figsize=(15, 10))
plot_idx = 1

for dtype, dpath in dataset_types.items():
    print(f"\nProcessing {dtype} dataset from: {dpath}")

    # Get a list of all .tif image file paths
    all_image_files = glob(os.path.join(dpath, '**', '*.tif'), recursive=True)
    print(f"Found {len(all_image_files)} .tif files in {dtype} set.")

    if not all_image_files:
        print(f"No .tif files found in {dtype} directory: {dpath}. Skipping.")
        continue

    # Select a small, representative sample of image paths
    # Ensure we don't try to sample more than available files
    sampled_images = np.random.choice(all_image_files, min(sample_size, len(all_image_files)), replace=False)

    for i, image_file in enumerate(sampled_images):
        try:
            with rasterio.open(image_file) as src:
                # Extract properties
                width = src.width
                height = src.height
                channels = src.count
                data_type = src.dtypes[0] if src.dtypes else 'unknown'

                properties = {
                    'filename': os.path.basename(image_file),
                    'dimensions': f"{width}x{height}",
                    'channels': channels,
                    'data_type': data_type
                }
                image_properties[dtype].append(properties)

                # Visualize the image
                ax = plt.subplot(len(dataset_types), sample_size, plot_idx)
                visualize_image(image_file, f"{dtype.capitalize()}: {os.path.basename(image_file)}", ax)
                plot_idx += 1

        except rasterio.errors.RasterioIOError as e:
            print(f"Error opening or reading image {image_file}: {e}")
        except Exception as e:
            print(f"An unexpected error occurred with image {image_file}: {e}")

plt.tight_layout()
plt.show()

# 5. Print the collected image properties for each dataset
print("\n--- Image Properties Summary ---")
for dtype, props_list in image_properties.items():
    if props_list:
        print(f"\n{dtype.capitalize()} Dataset:")
        for props in props_list:
            print(f"  Filename: {props['filename']}, Dimensions: {props['dimensions']}, Channels: {props['channels']}, Data Type: {props['data_type']}")
    else:
        print(f"\nNo properties collected for {dtype.capitalize()} dataset.")
```

Identify and analyze common image data issues such as varying image sizes, corrupted files, inconsistent color spaces, or unusual pixel value ranges within the unzipped images.
```python
import os
import rasterio
import pandas as pd
from glob import glob

# Define the base path to the extracted dataset and subdirectories
base_dataset_path = '/content/CloudSEN12_dataset/CloudSEN12_Dataset/'
train_path = os.path.join(base_dataset_path, 'train')
test_path = os.path.join(base_dataset_path, 'test')
val_path = os.path.join(base_dataset_path, 'val')

dataset_paths = {
    'train': train_path,
    'test': test_path, # Still include for completeness, even if empty
    'val': val_path
}

all_image_data = []
corrupted_files = []

print("Starting image property extraction...")

for dtype, dpath in dataset_paths.items():
    print(f"\nProcessing {dtype} dataset from: {dpath}")

    # Get a list of all .tif image file paths
    # The glob pattern needs to consider both images and masks are directly under subfolders like 'train/images' and 'train/masks'
    image_files = glob(os.path.join(dpath, '**', '*.tif'), recursive=True)
    print(f"Found {len(image_files)} .tif files in {dtype} set.")

    if not image_files:
        print(f"No .tif files found in {dtype} directory: {dpath}. Skipping property extraction for this set.")
        continue

    for image_file in image_files:
        try:
            with rasterio.open(image_file) as src:
                properties = {
                    'dataset_type': dtype,
                    'filepath': image_file,
                    'filename': os.path.basename(image_file),
                    'width': src.width,
                    'height': src.height,
                    'channels': src.count,
                    'data_type': src.dtypes[0] if src.dtypes else 'unknown'
                }
                all_image_data.append(properties)
        except rasterio.errors.RasterioIOError as e:
            print(f"Warning: Corrupted or unreadable file detected: {image_file} - {e}")
            corrupted_files.append({'filepath': image_file, 'error': str(e)})
        except Exception as e:
            print(f"An unexpected error occurred with file {image_file}: {e}")

print("\nImage property extraction complete.")

# Convert to DataFrame for easier analysis
if all_image_data:
    df_image_properties = pd.DataFrame(all_image_data)
    print("Collected image properties into DataFrame.")
    print("First 5 rows of image properties:")
    print(df_image_properties.head())
else:
    df_image_properties = pd.DataFrame(columns=['dataset_type', 'filepath', 'filename', 'width', 'height', 'channels', 'data_type'])
    print("No image properties collected.")


print("\n--- Analysis of Image Properties ---")

# 3a. Check for consistency in image dimensions
print("\n1. Image Dimensions Consistency:")
if not df_image_properties.empty:
    unique_dimensions = df_image_properties.groupby(['width', 'height']).size().reset_index(name='count')
    print(unique_dimensions)
    if len(unique_dimensions) == 1 and unique_dimensions.iloc[0]['width'] == 512 and unique_dimensions.iloc[0]['height'] == 512:
        print("All images have consistent dimensions of 512x512.")
    else:
        print("Image dimensions vary or are not all 512x512.")
else:
    print("No image data to analyze dimensions.")

# 3b. Analyze the distribution of channel counts
print("\n2. Channel Counts Distribution:")
if not df_image_properties.empty:
    channel_counts = df_image_properties['channels'].value_counts().sort_index()
    print(channel_counts)
    if set(channel_counts.index.values) == {1, 13}:
        print("All images consistently have either 1 or 13 channels.")
    else:
        print("Channel counts vary beyond 1 and 13.")
else:
    print("No image data to analyze channel counts.")

# 3c. Examine pixel data types and their distribution across different channel counts
print("\n3. Data Types by Channel Count:")
if not df_image_properties.empty:
    data_type_channel_dist = df_image_properties.groupby(['channels', 'data_type']).size().reset_index(name='count')
    print(data_type_channel_dist)
else:
    print("No image data to analyze data types.")

# 3d. Count the total number of images in each dataset, and count 1-channel and 13-channel images
print("\n4. Image Counts per Dataset and Channel Type:")
for dtype in dataset_paths.keys():
    if not df_image_properties.empty:
        dataset_df = df_image_properties[df_image_properties['dataset_type'] == dtype]
        total_count = len(dataset_df)
        channel_1_count = len(dataset_df[dataset_df['channels'] == 1])
        channel_13_count = len(dataset_df[dataset_df['channels'] == 13])
        other_channel_count = total_count - channel_1_count - channel_13_count

        print(f"  {dtype.capitalize()} Dataset: Total images = {total_count}")
        print(f"    - 1-channel images: {channel_1_count}")
        print(f"    - 13-channel images: {channel_13_count}")
        if other_channel_count > 0:
            print(f"    - Other channel counts: {other_channel_count}")
    else:
        print(f"  {dtype.capitalize()} Dataset: No image data.")

# 4. Identify potential data corruption
print("\n--- Corrupted Files ---")
if corrupted_files:
    print(f"Found {len(corrupted_files)} potentially corrupted or unreadable files:")
    for cf in corrupted_files:
        print(f"  - {cf['filepath']} (Error: {cf['error']})")
else:
    print("No corrupted or unreadable files identified.")

# Make the DataFrame available for subsequent steps
# globals()['df_image_properties'] = df_image_properties
```

```python
import os
import shutil
import pandas as pd
from sklearn.model_selection import train_test_split
from glob import glob

# Define the base path to the extracted dataset
base_dataset_path = '/content/CloudSEN12_dataset/CloudSEN12_Dataset/'

# Paths for original train and val data
original_train_path = os.path.join(base_dataset_path, 'train')
original_val_path = os.path.join(base_dataset_path, 'val')

# New directories for the re-split data
new_base_split_path = '/content/CloudSEN12_dataset/CloudSEN12_Dataset_Split/'
new_train_path = os.path.join(new_base_split_path, 'train')
new_val_path = os.path.join(new_base_split_path, 'val')
new_test_path = os.path.join(new_base_split_path, 'test')

# Create new directories if they don't exist and clear them if they do
for path in [new_train_path, new_val_path, new_test_path]:
    # Clear contents of the directory if it exists
    if os.path.exists(os.path.join(path, 'images')):
        shutil.rmtree(os.path.join(path, 'images'))
    if os.path.exists(os.path.join(path, 'masks')):
        shutil.rmtree(os.path.join(path, 'masks'))

    # Recreate the directories
    os.makedirs(os.path.join(path, 'images'), exist_ok=True)
    os.makedirs(os.path.join(path, 'masks'), exist_ok=True)

print(f"New split directories created and cleared at: {new_base_split_path}")

# Function to extract tile_id from filename (reusing from previous step)
def get_tile_id(filepath):
    filename = os.path.basename(filepath)
    if '_mask.tif' in filename:
        return filename.replace('_mask.tif', '')
    elif '.tif' in filename:
        return filename.replace('.tif', '')
    return None

# Collect all image and mask file paths from original train and val
all_image_files = []
all_mask_files = []

for dpath in [original_train_path, original_val_path]:
    all_image_files.extend(glob(os.path.join(dpath, 'images', '*.tif'), recursive=True))
    all_mask_files.extend(glob(os.path.join(dpath, 'masks', '*.tif'), recursive=True))

print(f"Found {len(all_image_files)} spectral images and {len(all_mask_files)} mask images in original datasets.")

# Create DataFrames for easier pairing
df_images = pd.DataFrame({'filepath': all_image_files})
df_images['tile_id'] = df_images['filepath'].apply(get_tile_id)

df_masks = pd.DataFrame({'filepath': all_mask_files})
df_masks['tile_id'] = df_masks['filepath'].apply(get_tile_id)

# Merge to get paired image and mask file paths
paired_data = pd.merge(df_images, df_masks, on='tile_id', suffixes=('_image', '_mask'))

print(f"Found {len(paired_data)} paired image-mask samples.")

# Split the paired data into new train, validation, and test sets
# First, split into train/test
train_val_pairs, test_pairs = train_test_split(paired_data, test_size=0.15, random_state=42)

# Then, split train_val into new train/val
train_pairs, val_pairs = train_test_split(train_val_pairs, test_size=0.15/0.85, random_state=42) # Adjust test_size for correct proportion

print(f"New split counts: Train={len(train_pairs)}, Validation={len(val_pairs)}, Test={len(test_pairs)}")

# Function to move files to new split directories
def move_files(dataframe, target_image_dir, target_mask_dir):
    for index, row in dataframe.iterrows():
        # Move image
        shutil.copy(row['filepath_image'], os.path.join(target_image_dir, os.path.basename(row['filepath_image'])))
        # Move mask
        shutil.copy(row['filepath_mask'], os.path.join(target_mask_dir, os.path.basename(row['filepath_mask'])))

print("Moving files to new 'train' directory...")
move_files(train_pairs, os.path.join(new_train_path, 'images'), os.path.join(new_train_path, 'masks'))

print("Moving files to new 'validation' directory...")
move_files(val_pairs, os.path.join(new_val_path, 'images'), os.path.join(new_val_path, 'masks'))

print("Moving files to new 'test' directory...")
move_files(test_pairs, os.path.join(new_test_path, 'images'), os.path.join(new_test_path, 'masks'))

print("Dataset re-splitting and file movement complete!")

# Update base_dataset_path to point to the new split for subsequent steps
globals()['base_dataset_path'] = new_base_split_path

# Confirm the contents of the new split directories
print("\n--- Confirming new split directory contents ---")
print(f"New train images: {len(os.listdir(os.path.join(new_train_path, 'images')))}")
print(f"New train masks: {len(os.listdir(os.path.join(new_train_path, 'masks')))}")
print(f"New validation images: {len(os.listdir(os.path.join(new_val_path, 'images')))}")
print(f"New validation masks: {len(os.listdir(os.path.join(new_val_path, 'masks')))}")
print(f"New test images: {len(os.listdir(os.path.join(new_test_path, 'images')))}")
print(f"New test masks: {len(os.listdir(os.path.join(new_test_path, 'masks')))}")
```

### Subtask 2: Standardize Image Dimensions and Normalize Pixel Values

Now that the datasets are correctly split, we need to process the images:
1.  Resize any `2048x2048` images to `512x512`.
2.  Normalize the `uint16` spectral images to a `0-1` float range.
3.  Ensure `uint8` masks are ready for use as class labels.

```python
import os
import rasterio
import numpy as np
from glob import glob
from tqdm import tqdm
from PIL import Image

# Assuming base_dataset_path is updated to the new split path from previous step
# base_dataset_path = '/content/CloudSEN12_dataset/CloudSEN12_Dataset_Split/'

output_base_path = '/content/CloudSEN12_dataset/CloudSEN12_Processed/'

# Create output directories
for subset in ['train', 'val', 'test']:
    os.makedirs(os.path.join(output_base_path, subset, 'images'), exist_ok=True)
    os.makedirs(os.path.join(output_base_path, subset, 'masks'), exist_ok=True)

print(f"Processed data will be saved to: {output_base_path}")

def process_and_save_image(filepath, output_dir, target_size=(512, 512), is_mask=False):
    try:
        with rasterio.open(filepath) as src:
            # Read all bands into a NumPy array
            image_data = src.read()

            # Handle resizing if necessary
            if image_data.shape[1] != target_size[0] or image_data.shape[2] != target_size[1]:
                # Rasterio read returns (bands, height, width), PIL expects (height, width, bands)
                # Resize using PIL for convenience, then convert back to (bands, height, width)
                # Handle multi-band images for resizing
                if image_data.ndim == 3:
                    # For multi-channel images, iterate through bands or convert to HWC for PIL
                    resized_bands = []
                    for band_idx in range(image_data.shape[0]):
                        band_img = Image.fromarray(image_data[band_idx, :, :])
                        resized_bands.append(np.array(band_img.resize(target_size, Image.NEAREST if is_mask else Image.BICUBIC)))
                    image_data = np.stack(resized_bands, axis=0)
                elif image_data.ndim == 2:
                    # For single-channel images
                    band_img = Image.fromarray(image_data)
                    image_data = np.array(band_img.resize(target_size, Image.NEAREST if is_mask else Image.BICUBIC))
                    image_data = np.expand_dims(image_data, axis=0) # Add band dimension back
                else:
                    print(f"Warning: Unexpected image dimension for resizing: {filepath}")
                    return # Skip this file

            if not is_mask:
                # Normalize uint16 spectral data to 0-1 float32
                image_data = image_data.astype(np.float32) / (2**16 - 1) # Max value for uint16
            else:
                # Masks remain uint8, but ensure correct type for saving
                image_data = image_data.astype(np.uint8)

            # Prepare output file path
            output_filepath = os.path.join(output_dir, os.path.basename(filepath))

            # Get profile for saving. Update height, width, and dtype.
            profile = src.profile
            profile.update({
                'height': target_size[0],
                'width': target_size[1],
                'dtype': image_data.dtype,
                'count': image_data.shape[0] if image_data.ndim == 3 else 1 # Ensure count is correct
            })
            if 'transform' in profile: # Remove or update transform for resized image
                profile.pop('transform') # Simplest is to remove if not correctly updating
            if 'crs' in profile: # Remove crs if not correctly updating
                profile.pop('crs')
            if 'affine' in profile: # Remove affine if not correctly updating
                profile.pop('affine')

            with rasterio.open(output_filepath, 'w', **profile) as dst:
                if image_data.ndim == 3:
                    for i in range(image_data.shape[0]):
                        dst.write(image_data[i, :, :], i + 1)
                elif image_data.ndim == 2: # For single band images
                    dst.write(image_data, 1)
                else:
                    print(f"Error saving: Unexpected image data dimensions {image_data.shape} for {output_filepath}")


    except rasterio.errors.RasterioIOError as e:
        print(f"Error opening or reading image {filepath}: {e}")
    except Exception as e:
        print(f"An unexpected error occurred with image {filepath}: {e}")


# Process each dataset subset
for subset in ['train', 'val', 'test']:
    current_image_dir = os.path.join(base_dataset_path, subset, 'images')
    current_mask_dir = os.path.join(base_dataset_path, subset, 'masks')

    output_image_dir = os.path.join(output_base_path, subset, 'images')
    output_mask_dir = os.path.join(output_base_path, subset, 'masks')

    print(f"\nProcessing {subset} images...")
    image_files = glob(os.path.join(current_image_dir, '*.tif'))
    for img_file in tqdm(image_files, desc=f"Processing {subset} spectral images"):
        process_and_save_image(img_file, output_image_dir, is_mask=False)

    print(f"Processing {subset} masks...")
    mask_files = glob(os.path.join(current_mask_dir, '*.tif'))
    for msk_file in tqdm(mask_files, desc=f"Processing {subset} masks"):
        process_and_save_image(msk_file, output_mask_dir, is_mask=True)

print("\nData preparation (resizing and normalization) complete!")

# Update base_dataset_path to point to the processed data for subsequent steps
globals()['base_dataset_path'] = output_base_path

# Optional: Verify a sample processed image
print("\nVerifying a sample processed image:")
if len(glob(os.path.join(output_base_path, 'train', 'images', '*.tif'))) > 0:
    sample_img_path = glob(os.path.join(output_base_path, 'train', 'images', '*.tif'))[0]
    with rasterio.open(sample_img_path) as src_processed:
        print(f"  Sample Image: {os.path.basename(sample_img_path)}")
        print(f"    Dimensions: {src_processed.width}x{src_processed.height}")
        print(f"    Channels: {src_processed.count}")
        print(f"    Data Type: {src_processed.dtypes[0]}")
        sample_data = src_processed.read(1)
        print(f"    Min Pixel Value (Band 1): {sample_data.min():.4f}")
        print(f"    Max Pixel Value (Band 1): {sample_data.max():.4f}")
```

The model should be configured to accept 13-channel input images and produce a 1-channel output (representing the segmentation mask/class labels). This will involve defining the encoder (downsampling path) and decoder (upsampling path) with convolutional layers, activation functions (e.g., ReLU), pooling layers (e.g., MaxPool2d), and upsampling layers (e.g., ConvTranspose2d) or interpolation.


**Reasoning**:
I need to implement the U-Net architecture as described in the subtask. This involves defining a `DoubleConv` helper module, the `UNet` class with its encoder and decoder paths, handling skip connections, and finally instantiating and printing a summary of the model.



```python
import torch
import torch.nn as nn

# 1. Define a DoubleConv helper module
class DoubleConv(nn.Module):
    def __init__(self, in_channels, out_channels):
        super(DoubleConv, self).__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(in_channels, out_channels, kernel_size=3, padding=1, bias=False),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(inplace=True),
            nn.Conv2d(out_channels, out_channels, kernel_size=3, padding=1, bias=False),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(inplace=True)
        )

    def forward(self, x):
        return self.conv(x)

# 2. Define the UNet class
class UNet(nn.Module):
    def __init__(self, in_channels=13, out_channels=1):
        super(UNet, self).__init__()
        self.in_channels = in_channels
        self.out_channels = out_channels

        # Encoder (Downsampling path)
        self.down_conv1 = DoubleConv(in_channels, 64)
        self.maxpool1 = nn.MaxPool2d(kernel_size=2, stride=2)
        self.down_conv2 = DoubleConv(64, 128)
        self.maxpool2 = nn.MaxPool2d(kernel_size=2, stride=2)
        self.down_conv3 = DoubleConv(128, 256)
        self.maxpool3 = nn.MaxPool2d(kernel_size=2, stride=2)
        self.down_conv4 = DoubleConv(256, 512)
        self.maxpool4 = nn.MaxPool2d(kernel_size=2, stride=2)

        # Bottleneck
        self.bottleneck = DoubleConv(512, 1024)

        # Decoder (Upsampling path)
        self.up_trans1 = nn.ConvTranspose2d(1024, 512, kernel_size=2, stride=2)
        self.up_conv1 = DoubleConv(1024, 512) # 512 from up_trans1 + 512 from skip connection
        self.up_trans2 = nn.ConvTranspose2d(512, 256, kernel_size=2, stride=2)
        self.up_conv2 = DoubleConv(512, 256) # 256 from up_trans2 + 256 from skip connection
        self.up_trans3 = nn.ConvTranspose2d(256, 128, kernel_size=2, stride=2)
        self.up_conv3 = DoubleConv(256, 128) # 128 from up_trans3 + 128 from skip connection
        self.up_trans4 = nn.ConvTranspose2d(128, 64, kernel_size=2, stride=2)
        self.up_conv4 = DoubleConv(128, 64) # 64 from up_trans4 + 64 from skip connection

        # Output layer
        self.out_layer = nn.Conv2d(64, out_channels, kernel_size=1)

    def forward(self, x):
        # Encoder
        x1 = self.down_conv1(x) # 13 -> 64
        x = self.maxpool1(x1) # 64
        x2 = self.down_conv2(x) # 64 -> 128
        x = self.maxpool2(x2) # 128
        x3 = self.down_conv3(x) # 128 -> 256
        x = self.maxpool3(x3) # 256
        x4 = self.down_conv4(x) # 256 -> 512
        x = self.maxpool4(x4) # 512

        # Bottleneck
        x = self.bottleneck(x) # 512 -> 1024

        # Decoder
        x = self.up_trans1(x) # 1024 -> 512 (upsample)
        # Skip connection: concatenate with corresponding encoder output
        x = torch.cat([x4, x], dim=1) # 512 + 512 = 1024
        x = self.up_conv1(x) # 1024 -> 512

        x = self.up_trans2(x) # 512 -> 256 (upsample)
        x = torch.cat([x3, x], dim=1) # 256 + 256 = 512
        x = self.up_conv2(x) # 512 -> 256

        x = self.up_trans3(x) # 256 -> 128 (upsample)
        x = torch.cat([x2, x], dim=1) # 128 + 128 = 256
        x = self.up_conv3(x) # 256 -> 128

        x = self.up_trans4(x) # 128 -> 64 (upsample)
        x = torch.cat([x1, x], dim=1) # 64 + 64 = 128
        x = self.up_conv4(x) # 128 -> 64

        # Output
        out = self.out_layer(x) # 64 -> 1
        return out

# 3. Instantiate the UNet model
model = UNet(in_channels=13, out_channels=1)

# Determine the device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)
print(f"Model moved to device: {device}")

# Print a summary of the model (optional, but useful for verification)
# Requires 'torchsummary' to be installed (pip install torchsummary)
try:
    from torchsummary import summary
    # Assuming input size of 512x512 for our images with 13 channels
    # Create a dummy input tensor and move it to the correct device
    dummy_input = torch.randn(1, 13, 512, 512).to(device)
    summary(model, input_size=(13, 512, 512), device=str(device))
except ImportError:
    print("torchsummary not installed. Skipping model summary.")
    print(model)

print("U-Net model architecture defined and instantiated successfully.")
```

### Implementing Multi-Class Dice Loss

```python
import torch.nn.functional as F

class DiceLoss(nn.Module):
    def __init__(self, num_classes, ignore_index=None, smooth=1e-6):
        super(DiceLoss, self).__init__()
        self.num_classes = num_classes
        self.ignore_index = ignore_index
        self.smooth = smooth

    def forward(self, inputs, targets):
        # inputs: (N, C, H, W) - raw logits
        # targets: (N, H, W) - class indices

        # Apply softmax to get probabilities for each class
        inputs = F.softmax(inputs, dim=1)  # (N, C, H, W)

        # Create one-hot encoded ground truth masks
        # targets need to be (N, 1, H, W) for one_hot, then squeeze to (N, H, W) again
        # one_hot returns (N, H, W, C), so permute to (N, C, H, W)
        targets_one_hot = F.one_hot(targets, num_classes=self.num_classes).permute(0, 3, 1, 2).float()

        # Flatten label and prediction tensors for class-wise calculation
        inputs_flat = inputs.view(-1, self.num_classes)
        targets_flat = targets_one_hot.view(-1, self.num_classes)

        loss = 0.0
        for c in range(self.num_classes):
            if self.ignore_index is not None and c == self.ignore_index:
                continue

            # Consider only pixels where target is not ignore_index for the current class
            # This filtering should be done *before* flattening if targets have ignore_index
            # More robust approach for ignore_index: mask out ignored pixels after one-hot encoding
            # For Dice Loss, usually you'd compute it on all classes and then average.
            # Let's adjust the one-hot to specifically handle ignore_index, by making it all zeros
            # for those positions.

            # A simpler way to handle ignore_index for Dice Loss is to exclude those pixels
            # from the calculation. For one-hot encoding, targets == ignore_index will result
            # in all zeros for that pixel across all channels, which naturally excludes it.

            # Get predictions and targets for current class
            input_c = inputs_flat[:, c]
            target_c = targets_flat[:, c]

            intersection = (input_c * target_c).sum()
            union = input_c.sum() + target_c.sum()

            dice_coefficient = (2. * intersection + self.smooth) / (union + self.smooth)
            loss += (1. - dice_coefficient)

        # Average the loss across classes. We divide by num_classes, excluding ignore_index if present.
        num_valid_classes = self.num_classes - (1 if self.ignore_index is not None else 0)
        return loss / num_valid_classes

print("DiceLoss class defined.")
```

```python
import torch
import torch.nn as nn
import numpy as np
import os
import rasterio
from torch.utils.data import Dataset, DataLoader
from glob import glob
import albumentations as A
from albumentations.pytorch import ToTensorV2
import matplotlib.pyplot as plt

# --- Part 1: U-Net Model Definition (from cell 56166eb6) ---
# 1. Define a DoubleConv helper module
class DoubleConv(nn.Module):
    def __init__(self, in_channels, out_channels):
        super(DoubleConv, self).__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(in_channels, out_channels, kernel_size=3, padding=1, bias=False),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(inplace=True),
            nn.Conv2d(out_channels, out_channels, kernel_size=3, padding=1, bias=False),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(inplace=True)
        )

    def forward(self, x):
        return self.conv(x)

# 2. Define the UNet class
class UNet(nn.Module):
    def __init__(self, in_channels=13, out_channels=1):
        super(UNet, self).__init__()
        self.in_channels = in_channels
        self.out_channels = out_channels

        # Encoder (Downsampling path)
        self.down_conv1 = DoubleConv(in_channels, 64)
        self.maxpool1 = nn.MaxPool2d(kernel_size=2, stride=2)
        self.down_conv2 = DoubleConv(64, 128)
        self.maxpool2 = nn.MaxPool2d(kernel_size=2, stride=2)
        self.down_conv3 = DoubleConv(128, 256)
        self.maxpool3 = nn.MaxPool2d(kernel_size=2, stride=2)
        self.down_conv4 = DoubleConv(256, 512)
        self.maxpool4 = nn.MaxPool2d(kernel_size=2, stride=2)

        # Bottleneck
        self.bottleneck = DoubleConv(512, 1024)

        # Decoder (Upsampling path)
        self.up_trans1 = nn.ConvTranspose2d(1024, 512, kernel_size=2, stride=2)
        self.up_conv1 = DoubleConv(1024, 512) # 512 from up_trans1 + 512 from skip connection
        self.up_trans2 = nn.ConvTranspose2d(512, 256, kernel_size=2, stride=2)
        self.up_conv2 = DoubleConv(512, 256) # 256 from up_trans2 + 256 from skip connection
        self.up_trans3 = nn.ConvTranspose2d(256, 128, kernel_size=2, stride=2)
        self.up_conv3 = DoubleConv(256, 128) # 128 from up_trans3 + 128 from skip connection
        self.up_trans4 = nn.ConvTranspose2d(128, 64, kernel_size=2, stride=2)
        self.up_conv4 = DoubleConv(128, 64) # 64 from up_trans4 + 64 from skip connection

        # Output layer
        self.out_layer = nn.Conv2d(64, out_channels, kernel_size=1)

    def forward(self, x):
        # Encoder
        x1 = self.down_conv1(x) # 13 -> 64
        x = self.maxpool1(x1) # 64
        x2 = self.down_conv2(x) # 64 -> 128
        x = self.maxpool2(x2) # 128
        x3 = self.down_conv3(x) # 128 -> 256
        x = self.maxpool3(x3) # 256
        x4 = self.down_conv4(x) # 256 -> 512
        x = self.maxpool4(x4) # 512

        # Bottleneck
        x = self.bottleneck(x) # 512 -> 1024

        # Decoder
        x = self.up_trans1(x) # 1024 -> 512 (upsample)
        # Skip connection: concatenate with corresponding encoder output
        x = torch.cat([x4, x], dim=1) # 512 + 512 = 1024
        x = self.up_conv1(x) # 1024 -> 512

        x = self.up_trans2(x) # 512 -> 256 (upsample)
        x = torch.cat([x3, x], dim=1) # 256 + 256 = 512
        x = self.up_conv2(x) # 512 -> 256

        x = self.up_trans3(x) # 256 -> 128 (upsample)
        x = torch.cat([x2, x], dim=1) # 128 + 128 = 256
        x = self.up_conv3(x) # 256 -> 128

        x = self.up_trans4(x) # 128 -> 64 (upsample)
        x = torch.cat([x1, x], dim=1) # 64 + 64 = 128
        x = self.up_conv4(x) # 128 -> 64

        # Output
        out = self.out_layer(x) # 64 -> 1
        return out

print("U-Net model architecture defined.")

# --- Part 2: CloudSEN12Dataset and DataLoader definitions ---

class CloudSEN12Dataset(Dataset):
    def __init__(self, root_dir, subset, transform=None):
        self.root_dir = root_dir
        self.subset_path = os.path.join(root_dir, subset)
        self.transform = transform

        self.image_files = sorted(glob(os.path.join(self.subset_path, 'images', '*.tif')))
        self.mask_files = sorted(glob(os.path.join(self.subset_path, 'masks', '*.tif')))

        # Ensure images and masks are correctly paired. Assuming filenames are identical except for '_mask.tif'
        paired_data = []
        image_basenames = {os.path.basename(f).replace('.tif', ''): f for f in self.image_files}
        mask_basenames = {os.path.basename(f).replace('_mask.tif', ''): f for f in self.mask_files}

        for img_id, img_path in image_basenames.items():
            if img_id in mask_basenames:
                paired_data.append((img_path, mask_basenames[img_id]))

        self.paired_data = sorted(paired_data)

        if not self.paired_data:
            print(f"Warning: No paired image-mask data found for subset {subset} in {root_dir}")

    def __len__(self):
        return len(self.paired_data)

    def __getitem__(self, idx):
        img_path, mask_path = self.paired_data[idx]

        # Load image (spectral data)
        with rasterio.open(img_path) as src_img:
            image = src_img.read().astype(np.float32) # Already normalized to 0-1 from previous step

        # Load mask (labels)
        with rasterio.open(mask_path) as src_mask:
            mask = src_mask.read(1).astype(np.int64) # Ensure mask is int64 for PyTorch loss functions

        # Rasterio reads (C, H, W). Albumentations expects (H, W, C) for images and (H, W) for masks.
        # Permute image to (H, W, C)
        image = np.transpose(image, (1, 2, 0))

        if self.transform:
            # Apply transform to both image and mask
            augmented = self.transform(image=image, mask=mask)
            image = augmented['image']
            mask = augmented['mask']

        # Albumentations ToTensorV2 converts (H,W,C) to (C,H,W) and normalizes if input was uint8.
        # Since our image is float32 0-1, it just permutes.
        # Mask needs to be unsqueezed to have a channel dimension (1, H, W) for compatibility.
        mask = mask.unsqueeze(0) # Add channel dimension for mask

        return image, mask

# Define augmentation pipeline for training data
train_transform = A.Compose([
    A.HorizontalFlip(p=0.5),
    A.VerticalFlip(p=0.5),
    # A.Rotate(limit=45, interpolation=0, p=0.5), # Interpolation=0 for nearest neighbor with masks
    ToTensorV2() # Converts image to PyTorch tensor (C, H, W) and normalizes if uint8
])

# No augmentation for validation and test data, just conversion to tensor
val_test_transform = A.Compose([
    ToTensorV2()
])

# Instantiate CloudSEN12Dataset for each subset
processed_data_root = '/content/CloudSEN12_dataset/CloudSEN12_Processed/' # Ensure this path is correct

train_dataset = CloudSEN12Dataset(processed_data_root, 'train', transform=train_transform)
val_dataset = CloudSEN12Dataset(processed_data_root, 'val', transform=val_test_transform)
test_dataset = CloudSEN12Dataset(processed_data_root, 'test', transform=val_test_transform)

print(f"Train dataset size: {len(train_dataset)}")
print(f"Validation dataset size: {len(val_dataset)}")
print(f"Test dataset size: {len(test_dataset)}")

# Create DataLoader instances
batch_size = 4
num_workers = 2

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=num_workers)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False, num_workers=num_workers)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False, num_workers=num_workers)

print("Data loading pipeline setup complete.")


# --- Part 3: Evaluation Metrics and Loop ---

# Define the ignore index based on observation
IGNORE_INDEX = 99

# 2. Re-calculate all_unique_mask_values, actual_class_labels, and num_actual_classes
all_unique_mask_values = set()

print("\nCollecting unique mask values from train dataset...")
for i in range(len(train_dataset)):
    _, mask = train_dataset[i]
    unique_values = np.unique(mask.numpy())
    all_unique_mask_values.update(unique_values)

print("Collecting unique mask values from validation dataset...")
for i in range(len(val_dataset)):
    _, mask = val_dataset[i]
    unique_values = np.unique(mask.numpy())
    all_unique_mask_values.update(unique_values)

# Check if the minimum class label is 0
if 0 not in all_unique_mask_values:
    print(f"Warning: Class 0 not found in mask values. Unique values: {sorted(list(all_unique_mask_values))}")

print(f"Detected unique mask values (class labels): {sorted(list(all_unique_mask_values))}")

# Filter out the IGNORE_INDEX from the unique mask values to determine actual classes
actual_class_labels = [val for val in all_unique_mask_values if val != IGNORE_INDEX]

# Determine the highest valid class label. If labels are 0,1,2, then max is 2, so 3 classes.
if actual_class_labels:
    num_actual_classes = int(max(actual_class_labels)) + 1
else:
    num_actual_classes = 1 # Fallback, if only ignore index or no labels are found

print(f"Detected actual class labels (excluding ignore_index={IGNORE_INDEX}): {sorted(actual_class_labels)}")
print(f"Setting final_out_channels for UNet to: {num_actual_classes}")

# 3. Instantiate the UNet model with correct output channels
model = UNet(in_channels=13, out_channels=num_actual_classes)

# 4. Move the model to the appropriate device (GPU if available, else CPU)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)
print(f"Model moved to device: {device}")

# 5. Load the saved model weights
model_save_path = 'best_unet_model.pth'
if os.path.exists(model_save_path):
    model.load_state_dict(torch.load(model_save_path, map_location=device))
    print(f"Loaded model weights from {model_save_path}")
else:
    print(f"Error: Model weights not found at {model_save_path}. Please ensure the model was trained and saved.")

# 6. Set the model to evaluation mode
model.eval()
print("Model set to evaluation mode.")

# 7. Define functions to calculate accuracy, precision, recall, and F1-score
def calculate_accuracy(preds, targets, ignore_index=None):
    # preds: (N, H, W) class indices
    # targets: (N, H, W) class indices
    if ignore_index is not None:
        valid_pixels = (targets != ignore_index)
        preds = preds[valid_pixels]
        targets = targets[valid_pixels]
    if targets.numel() == 0:
        return float('nan') # No valid pixels to calculate accuracy
    correct = (preds == targets).sum().item()
    total = targets.numel()
    return correct / total

def calculate_precision_recall_f1(preds, targets, num_classes, ignore_index=None, average='macro'):
    # preds: (N, H, W) class indices
    # targets: (N, H, W) class indices

    if ignore_index is not None:
        valid_pixels = (targets != ignore_index)
        preds = preds[valid_pixels]
        targets = targets[valid_pixels]

    if targets.numel() == 0:
        return float('nan'), float('nan'), float('nan')

    # Convert to one-hot for F1, precision, recall calculation per class
    # Remove the ignore_index from num_classes for this part if it's the highest label
    # Here we are iterating up to num_classes which already excludes ignore_index

    true_positives = torch.zeros(num_classes, dtype=torch.long, device=device)
    false_positives = torch.zeros(num_classes, dtype=torch.long, device=device)
    false_negatives = torch.zeros(num_classes, dtype=torch.long, device=device)

    for cls_id in range(num_classes):
        if cls_id == ignore_index: # Ensure ignore_index is not considered as a class here
            continue

        pred_mask = (preds == cls_id)
        target_mask = (targets == cls_id)

        true_positives[cls_id] = (pred_mask & target_mask).sum().item()
        false_positives[cls_id] = (pred_mask & ~target_mask).sum().item()
        false_negatives[cls_id] = (~pred_mask & target_mask).sum().item()

    precision = []
    recall = []
    f1 = []

    for cls_id in range(num_classes):
        if cls_id == ignore_index:
            continue

        tp = true_positives[cls_id].item()
        fp = false_positives[cls_id].item()
        fn = false_negatives[cls_id].item()

        if (tp + fp) > 0: # Avoid division by zero
            p = tp / (tp + fp)
        else:
            p = float('nan')
        precision.append(p)

        if (tp + fn) > 0: # Avoid division by zero
            r = tp / (tp + fn)
        else:
            r = float('nan')
        recall.append(r)

        if (p + r) > 0 and not np.isnan(p) and not np.isnan(r): # Avoid division by zero and NaN in sum
            f = 2 * (p * r) / (p + r)
        else:
            f = float('nan')
        f1.append(f)

    # Filter out NaN values before averaging
    valid_precision = [p for p in precision if not np.isnan(p)]
    valid_recall = [r for r in recall if not np.isnan(r)]
    valid_f1 = [f for f in f1 if not np.isnan(f)]

    if average == 'macro':
        macro_precision = np.mean(valid_precision) if valid_precision else 0.0
        macro_recall = np.mean(valid_recall) if valid_recall else 0.0
        macro_f1 = np.mean(valid_f1) if valid_f1 else 0.0
        return macro_precision, macro_recall, macro_f1
    else:
        # For 'weighted' or 'micro' average, further logic would be needed
        # For simplicity, returning macro average
        return macro_precision, macro_recall, macro_f1

# Re-define calculate_iou (copied from previous code block for completeness)
def calculate_iou(preds, targets, num_classes, ignore_index=None):
    # preds: (N, C, H, W) raw logits
    # targets: (N, H, W) class indices (already squeezed)

    # Convert predictions to class indices (N, H, W)
    preds = torch.argmax(preds, dim=1)

    # Flatten for easier comparison
    preds = preds.flatten()
    targets = targets.flatten()

    iou_list = []
    for cls_id in range(num_classes):
        if ignore_index is not None and cls_id == ignore_index: # Skip ignore index for IoU calculation
            continue

        pred_mask = (preds == cls_id)
        target_mask = (targets == cls_id)

        # Filter out pixels that are part of the ignore_index in the target
        if ignore_index is not None:
            valid_pixels = (targets != ignore_index)
            pred_mask = pred_mask[valid_pixels]
            target_mask = target_mask[valid_pixels]

        # If after filtering, no valid pixels remain for this class, skip it
        if target_mask.sum() == 0 and pred_mask.sum() == 0: # Both are empty for this class (after filtering ignore index)
            iou_list.append(float('nan')) # Or 0.0, if you prefer.
            continue

        intersection = (pred_mask & target_mask).sum().item()
        union = (pred_mask | target_mask).sum().item()

        if union == 0:
            iou_list.append(float('nan')) # Class not present in ground truth or predictions (after ignore_index filtering)
        else:
            iou_list.append(intersection / union)

    # Filter out NaN values (classes not present) before calculating mean IoU
    valid_iou = [iou for iou in iou_list if not np.isnan(iou)]
    if len(valid_iou) == 0:
        return 0.0 # No valid classes to compute IoU for
    return np.mean(valid_iou)


print("Evaluation metrics functions (Accuracy, Precision, Recall, F1, IoU) defined.")

# 8. Iterate through the test_loader
total_iou = 0.0
total_accuracy = 0.0
total_precision = 0.0
total_recall = 0.0
total_f1 = 0.0
num_batches = 0

print("\nEvaluating model on the test set...")
with torch.no_grad(): # Disable gradient calculations
    for images, masks in test_loader:
        images = images.to(device)
        masks = masks.to(device)
        masks = masks.squeeze(1) # Squeeze the mask channel

        outputs = model(images)
        preds = torch.argmax(outputs, dim=1) # Convert logits to class labels

        # Aggregate metrics
        total_iou += calculate_iou(outputs, masks, num_actual_classes, ignore_index=IGNORE_INDEX) * images.size(0)
        total_accuracy += calculate_accuracy(preds, masks, ignore_index=IGNORE_INDEX) * images.size(0)

        batch_precision, batch_recall, batch_f1 = calculate_precision_recall_f1(preds, masks, num_actual_classes, ignore_index=IGNORE_INDEX, average='macro')
        total_precision += batch_precision * images.size(0)
        total_recall += batch_recall * images.size(0)
        total_f1 += batch_f1 * images.size(0)

        num_batches += images.size(0)

# 9. Print overall average values
if num_batches > 0:
    avg_iou = total_iou / num_batches
    avg_accuracy = total_accuracy / num_batches
    avg_precision = total_precision / num_batches
    avg_recall = total_recall / num_batches
    avg_f1 = total_f1 / num_batches

    print(f"\n--- Test Set Evaluation Results ({num_batches} samples) ---")
    print(f"Average IoU: {avg_iou:.4f}")
    print(f"Average Accuracy: {avg_accuracy:.4f}")
    print(f"Average Precision (Macro): {avg_precision:.4f}")
    print(f"Average Recall (Macro): {avg_recall:.4f}")
    print(f"Average F1-score (Macro): {avg_f1:.4f}")
else:
    print("No test samples available for evaluation.")

# 10. Select a few sample images from the test_dataset for visualization
num_samples_to_visualize = min(5, len(test_dataset))
sample_indices = np.random.choice(len(test_dataset), num_samples_to_visualize, replace=False)

print(f"\nVisualizing {num_samples_to_visualize} sample predictions from the test set...")

plt.figure(figsize=(15, num_samples_to_visualize * 5))
for i, idx in enumerate(sample_indices):
    image, gt_mask = test_dataset[idx]

    # Move image to device and add batch dimension for prediction
    image_for_pred = image.unsqueeze(0).to(device)

    with torch.no_grad():
        output = model(image_for_pred)
        predicted_mask = torch.argmax(output, dim=1).squeeze(0).cpu().numpy()

    # Move original image and ground truth mask back to CPU and to numpy for plotting
    original_image_np = image.permute(1, 2, 0).cpu().numpy() # (C,H,W) -> (H,W,C)
    gt_mask_np = gt_mask.squeeze(0).cpu().numpy()

    # Visualize original spectral image (RGB composite), ground truth, and predicted mask
    ax = plt.subplot(num_samples_to_visualize, 3, i * 3 + 1)
    # Assuming 13-channel images have RGB bands at indices 1, 2, 3 (or similar)
    # For Sentinel-2, often B4, B3, B2 for true color, which might be indices 3, 2, 1 in a 13-band array
    # Let's use the first 3 channels for a generic visualization
    display_img = original_image_np[:, :, :3]
    # Normalize for display if needed (though it should be 0-1 already)
    display_img = (display_img - display_img.min()) / (display_img.max() - display_img.min())
    ax.imshow(display_img)
    ax.set_title("Original Image (RGB)")
    ax.axis('off')

    ax = plt.subplot(num_samples_to_visualize, 3, i * 3 + 2)
    ax.imshow(gt_mask_np, cmap='viridis') # Use a colormap for masks
    ax.set_title("Ground Truth Mask")
    ax.axis('off')

    ax = plt.subplot(num_samples_to_visualize, 3, i * 3 + 3)
    ax.imshow(predicted_mask, cmap='viridis') # Use same colormap for consistency
    ax.set_title("Predicted Mask")
    ax.axis('off')

plt.tight_layout()
plt.show()
```
